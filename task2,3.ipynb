```python
import requests
import csv
```

```python
data=requests.get("https://www.metaweather.com/api/location/1521894/")
```

```python
weather=data.json()
weather['consolidated_weather']
```

```python
with open("weather.csv","w",newline="")as h:
    writer=csv.DictWriter(h,fieldnames=["applicable_date","the_temp","wind_speed","air_pressure","humidity","visibility"])
    writer.writeheader()
    for i in weather['consolidated_weather']:
        datetime=i["applicable_date"]
        temperature=i["the_temp"]
        wind_speed=i["wind_speed"]
        air_pressure=i["air_pressure"]
        humidity=i["humidity"]
        visibility=i["visibility"]
        
        writer.writerow({
            "applicable_date":datetime,
            "the_temp":temperature,
            "wind_speed":wind_speed,
            "air_pressure":air_pressure,
            "humidity":humidity,
            "visibility":visibility




        })




  




 
```

```python
new_data=requests.get("https://jsonplaceholder.typicode.com/users")
users=new_data.json()
users
```

```python
with open("users_data.csv","w",newline="")as n:
    write=csv.DictWriter(n,fieldnames=["name","username","email","street","suite","city","zipcode","geo_lat","geo_lng"])
    write.writeheader()
    for i in users:
        name=i["name"]
        username=i["username"]
        email=i["email"]
        street=i["address"]["street"]
        suite=i["address"]["suite"]
        city=i["address"]["city"]
        z_code=i["address"]["zipcode"]
        geo_lat=i["address"]["geo"]["lat"]
        geo_long=i["address"]["geo"]["lng"]
        
        write.writerow({
            "name":name,
            "username":username,
            "email":email,
            "street":street,
            "suite":suite,
            "city":city,
            "zipcode":z_code,
            "geo_lat":geo_lat,
            "geo_lng":geo_long







        })
    
```
